NESTED LOOP

```python
limit=int(input('Enter number :'))
for num in range (2,limit+1):
  is_divisible=False
  k=2
  while k<=num//2:
    if num % k==0:
      is_divisible=True
      break;
    k+=1
  if not is_divisible:
    print(num,end=' ')
```

FINDING THE SQUARE OF A NUMBER

```python
def square(number):
  return pow(number,2)
s=square(5)
print(s)
```

prime in given range

```python
def isPrime(number):

    for factor in range(2, (number//2)+1):
        if number%factor == 0:

              return False

    return True

number = int(input('Enter the number '))
print(isPrime(number))
```

```python
def printPrimes(llimit, ulimit):
  for num in range(llimit,ulimit+1):
    if isPrime(num)==True:
      print(num,end=' ')
printPrimes(5,50)
```

swap 2 numbers

```python
def swap(x,y):
  t=x
  x=y
  y=t
  return x,y
a=5
b=7
a,b=swap(a,b)
print(a,b)
```
